```python
v = 3.0
c = 2.0
T = 1.0

import matplotlib.pyplot as plt
import math
import matplotlib.animation as anim

%matplotlib notebook

fig = plt.figure(figsize=(12,6))
ax = plt.axes(xlim=(-20,20), ylim=(-10,10))
circles = []
sources = []
t = 0.0
dt = 0.1
theSource = plt.Circle((v*t,0),0.1, color='gray', fill=True)
ax.add_patch(theSource)

def animate(frame):
    t = frame/200*5
    theSource.set_center((v*t,0))
    nbCercles = math.ceil(t/T)
    if nbCercles > len(circles) :
        newCircle = plt.Circle((v*t,0),0, color='red', fill=False)
        circles.append(newCircle)
        ax.add_patch(newCircle)
        newSource = plt.Circle((v*t,0),0.1, color='blue', fill=True)
        ax.add_patch(newSource)
        sources.append(newSource)
    numCercle = 0
    for aCircle in circles:
        aCircle.set_radius(c*(t-numCercle*T))
        numCercle+=1
    return theSource, circles, sources,

theAnimation = anim.FuncAnimation(fig, animate, frames=200, blit=False, interval=30, repeat=False)
plt.show()

```
